```python
import pandas as pd
```

```python
#1. Faça uma pandas Series com valores inteiros randomicos de tamanho 30 de numeros entre 5 e 15
serie_1 = pd.Series(np.random.randint(5,16,30))
```

```python
# Divida sua serie em duas partes e gere duas series novas 
serie_1_a = serie_1[0:15]
serie_1_b =  serie_1[15:30]
print(serie_1_a)
print(serie_1_b)
```

```python
    # Adicione 5 valores a cada serie e mantenha os dois elementos como series
serie_2_a = serie_1_a.append(pd.Series(np.random.randint(5,16,5)), ignore_index=True)
serie_2_b = serie_1_b.append(pd.Series(np.random.randint(5,16,5)), ignore_index=True)
print(serie_2_a)
print(serie_2_b)
```

```python
serie_2_a.unique()
```

```python
# Conte as aparições de cada elemento da Serie
print(serie_2_a.value_counts())
print(serie_2_b.value_counts())
```

```python
   # Faça comparações booleanas entre as das series
igualdade = (serie_2_a) == (serie_2_b)
maior_a = (serie_2_a) > (serie_2_b)
menor_a = (serie_2_a) < (serie_2_b)
```

```python
# Junte as duas series em um dataframe
lista_series = [serie_2_a,serie_2_b,igualdade,maior_a,menor_a]
df_1 = pd.concat(lista_series, axis=1)
df_1.columns = ['serie_a', 'serie_b', 'igualdade', 'a>b', 'a<b']
df_1
```
